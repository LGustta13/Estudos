- Separando em dados de treino e dados de teste

```python
from sklearn.model_selection import train_test_split
```

```python
x = table.drop('Vendas',axis = 1)
y = table['Vendas']
```

#### Temos um problema de regressão - Vamos escolher os modelos que vamos usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
import numpy as np

# Selecionando os dados para treino e teste
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)

# Treinando a AI
lin_reg = LinearRegression()
rf_reg = RandomForestRegressor()

lin_reg.fit(x_train,y_train)
rf_reg.fit(x_train,y_train)

# Testando a AI
test_pred_lin = lin_reg.predict(x_test)
test_pred_rf = rf_reg.predict(x_test)
```

#### Teste da AI e Avaliação do Melhor Modelo

- Vamos usar o R² -> diz o % que o nosso modelo consegue explicar o que acontece

```python
print(metrics.r2_score(y_test,test_pred_lin))
print(metrics.r2_score(y_test,test_pred_rf))

df_result = pd.DataFrame()
df_result['y_test'] = y_test
df_result['y_pred_lin'] = test_pred_lin
df_result['y_pred_rf'] = test_pred_rf
df_result = df_result.reset_index(drop=True)
```

#### Visualização Gráfica das Previsões

```python
fig = plt.figure(figsize=(15,5))
sns.lineplot(data = df_result)
plt.show()
display(df_result)

# pd.Dataframe(dados, linha, coluna, ...)
importance = pd.DataFrame(rf_reg.feature_importances_,x_train.columns,columns = ["Importance"])
plt.figure(figsize=(5,5))
sns.barplot(x=importance.index,y=importance["Importance"])
plt.show
```
